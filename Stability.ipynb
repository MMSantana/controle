# Symbolic Approach

```python
from sympy.solvers import solve
from sympy.solvers.solveset import nonlinsolve
from sympy.functions.elementary.exponential import exp
from sympy.functions.special.delta_functions import Heaviside
from sympy import Symbol
```

```python
S, E, I, G, R, D, t = Symbol('S'), Symbol('E'), Symbol('I'), Symbol('G'), Symbol('R'), Symbol('D'), Symbol('t')

beta, delta, alpha, gamma, rho, theta = 4/5, 0.2, 0.2, 0.2, 1/9, 0.1

N = S + E + I + G + R + D

nonlinsolve(
    [
        -beta * S * (I + E) / N + (-S*(0.3 + exp(-t-10)) + (N - D)/4*(0.3 + exp(-t-10)))*Heaviside(D - 5000),
        beta * S * (I + E) / N - delta * E + (-E*(0.3 + exp(-t-10)) + (N - D)/4*(0.3 + exp(-t-10)))*Heaviside(D - 5000),
        delta * E - (1 - alpha) * gamma * I - alpha * I * rho + (-I*(0.3 + exp(-t-10)) + (N - D)/4*(0.3 + exp(-t-10)))*Heaviside(D - 5000),
        alpha * I * rho - (1 - 0.3 + G / N) * gamma * G - (0.3 + G / N) * theta * G,
        (1 - alpha) * gamma * I + (1 - (0.3 + G / N)) * gamma * G + (-R*(0.3 + exp(-t-10)) + (N - D)/4*(0.3 + exp(-t-10)))*Heaviside(D - 5000),
        (0.3 + G / N) * theta * G
    ], [S, E, I, G, R, D, t]
    
)
```

# Numeric Approach

```python
import numpy as np
```

```python
from scipy.optimize import fsolve
import math

def equations(p):
    S, E, I, G, R, D = p
    
    beta, delta, alpha, gamma, rho, theta = 4/5, 0.2, 0.2, 0.2, 1/9, 0.1
    
    N = 1000000
    return  [-beta * S * (I + E) / N + (-S*(0.3) + (N - D)/4*(0.3))*np.heaviside(D - 5000),
            beta * S * (I + E) / N - delta * E + (-E*(0.3) + (N - D)/4*(0.3))*np.heaviside(D - 5000),
            delta * E - (1 - alpha) * gamma * I - alpha * I * rho + (-I*(0.3) + (N - D)/4*(0.3))*np.heaviside(D - 5000),
            alpha * I * rho - (1 - 0.3 + G / N) * gamma * G - (0.3 + G / N) * theta * G,
            (1 - alpha) * gamma * I + (1 - (0.3 + G / N)) * gamma * G + (-R*(0.3) + (N - D)/4*(0.3))*np.heaviside(D - 5000),
            (0.3 + G / N) * theta * G]

S, E, I, G, R, D =  fsolve(equations, (1000, 1000, 1000, 1000, 1000, 1000))

# print equations((x, y))
```
